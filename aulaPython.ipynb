```python
#importar bibliotecas
import pandas as pd
```

```python
#carregar datasets
# carrega dataset com infos sobre os pedidos: olist_orders_dataset.csv
olist_orders = pd.read_csv("olist_orders_dataset.csv")

# carrega dataset com infos sobre as avaliações dos pedidos: olist_order_reviews_dataset
reviws_orders = pd.read_csv("olist_order_reviews_dataset.csv")

# carrega dataset com infos sobre os pedidos: olist_order_payments_dataset.csv
information_orders = pd.read_csv("olist_order_payments_dataset.csv")

# carrega dataset do consumidor: olist_customers_dataset.csv
olist_consumers = pd.read_csv("olist_customers_dataset.csv")

# carrega dataset com os itens dos pedidos: olist_order_items_datase.csv
items_orders = pd.read_csv("olist_order_items_dataset.csv")

# carrega dataset com infos sobre os produtos: olist_products_dataset.csv
olist_products = pd.read_csv("olist_products_dataset.csv")

# carrega dataset sobre os vendedores: olist_sellers_dataset.csv
olist_sellers = pd.read_csv("olist_sellers_dataset.csv")

# carrega dataset com a tradução dos produtos:  product_category_name_translation
translation_products = pd.read_csv("product_category_name_translation.csv")
```

```python
#Merge dos datasets
#Merge de todos os datasets referentes ao consumidor
# A = olist_orders_dataset.csv, olist_order_reviews_dataset.csv, olist_order_payments_dataset.csv
# em order_id
# customer_df = A e olist_customers_dataset.csv em customer_id
```

```python
AA = pd.merge(olist_orders, reviws_orders, on='order_id', how='left')
A = pd.merge(AA, information_orders, on='order_id', how='left')
customer_df = pd.merge(A, olist_consumers, on='customer_id', how='left')
customer_df
```

```python
# Merge de todos os datasets referentes ao vendedor
# B = olist_order_items_dataset e olist_products_dataset em product_id
# C = B e olist_sellers_dataset em seller_id
# seller_df = C e product_category_name_translation.csv em product_category_name
```

```python
B = pd.merge(items_orders, olist_products, on='product_id', how='left')
C = pd.merge(B, olist_sellers, on='seller_id', how='left')
seller_df = pd.merge(C,translation_products, on='product_category_name', how='left')
seller_df
```

```python
#Merge dos datasets do consumidor e do vendedor
# df_ecommerce = merge do customer_df e do seller_df em order_id
df_ecommerce = pd.merge(customer_df,seller_df, on='order_id', how='left')
df_ecommerce
```
